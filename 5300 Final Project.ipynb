# Double Spring Pendulum with Damping

## Introduction

In this notebook, we simulate a double spring pendulum with:
- Two point masses, $m_1$ and $m_2$.
- Two massless springs with equilibrium lengths $l_1$ and $l_2$.
- Gravitational acceleration $g$.
- Damping force with damping coefficients $b_1$ and $b_2$.

### Equations of Motion

Let:
- $\theta_1$ be the angle of the first spring from the vertical (down is positive),
- $\theta_2$ be the angle of the second spring from the vertical.
- $r_1$ be the extension of the first spring from its equilibrium.
- $r_2$ be the extension of the second spring from its equilibrium.

# Rayleigh Dissipation Function

The **Rayleigh dissipation function** is a scalar function that models the energy dissipated due to non-conservative forces, such as friction or damping, in a mechanical system. It is particularly useful in the framework of Lagrangian mechanics to account for dissipative effects. For generalized coordinates $q_1, q_2, \dots , q_N$ with velocities $\dot q_i \,$, a Rayleigh dissipation function that represents *linear* (viscous) damping is

$$
R \;=\; \tfrac{1}{2}\;\sum_{i=1}^{N} c_i \,\dot q_i^{\,2},
$$

where $c_i>0$ is the viscous-damping coefficient associated with $\dot q_i$.


## Usage in Lagrangian Mechanics

In the presence of dissipative forces, the Euler–Lagrange equations are modified by including a generalized force term derived from the Rayleigh dissipation function. For each generalized coordinate $q_i$, the modified equation is:

$$
\frac{d}{dt} \left(\frac{\partial \mathcal{L}}{\partial \dot{q}_i}\right) - \frac{\partial \mathcal{L}}{\partial q_i} = -\frac{\partial R}{\partial \dot{q}_i},
$$

where $\mathcal{L}$ is the Lagrangian of the system.

For more detailed information, you can go to the [Wikipedia article on Rayleigh dissipation function](https://en.wikipedia.org/wiki/Rayleigh_dissipation_function).

The Euler–Lagrange equations with damping give the following coupled ODEs:

$$
\text{Euler–Lagrange Equation for }\theta_1
$$
\begin{aligned}
&(m_{1}+m_{2})(l_{1}+x_{1})^{2}\,\ddot{\theta}_{1}
+ m_{2}(l_{1}+x_{1})(l_{2}+x_{2})\cos(\theta_{1}-\theta_{2})\,\ddot{\theta}_{2}
- m_{2}(l_{1}+x_{1})\sin(\theta_{1}-\theta_{2})\,\ddot{x}_{2} \\[4pt]
&\quad
+ 2(m_{1}+m_{2})(l_{1}+x_{1})\,\dot{\theta}_{1}\dot{x}_{1}
+ 2m_{2}(l_{1}+x_{1})\cos(\theta_{1}-\theta_{2})\,\dot{\theta}_{2}\dot{x}_{2} \\[4pt]
&\quad
+ m_{2}(l_{1}+x_{1})(l_{2}+x_{2})\sin(\theta_{1}-\theta_{2})\,\dot{\theta}_{2}^{2}
- (m_{1}+m_{2})g(l_{1}+x_{1})\sin\theta_{1}=0
\end{aligned}



$$
\text{Euler–Lagrange Equation for }\theta_2
$$
\begin{aligned}
&m_{2}(l_{2}+x_{2})^{2}\,\ddot{\theta}_{2}
+ m_{2}(l_{1}+x_{1})(l_{2}+x_{2})\cos(\theta_{1}-\theta_{2})\,\ddot{\theta}_{1}
+ m_{2}(l_{2}+x_{2})\sin(\theta_{1}-\theta_{2})\,\ddot{x}_{1} \\[4pt]
&\quad
+ 2m_{2}(l_{2}+x_{2})\,\dot{\theta}_{2}\dot{x}_{2}
+ 2m_{2}(l_{2}+x_{2})\cos(\theta_{1}-\theta_{2})\,\dot{\theta}_{1}\dot{x}_{1} \\[4pt]
&\quad
- m_{2}(l_{1}+x_{1})(l_{2}+x_{2})\sin(\theta_{1}-\theta_{2})\,\dot{\theta}_{1}^{2}
- m_{2}g(l_{2}+x_{2})\sin\theta_{2}=0
\end{aligned}


$$
\text{Euler–Lagrange Equation for }x_1
$$
\begin{aligned}
&(m_{1}+m_{2})\,\ddot{x}_{1}
+ m_{2}\cos(\theta_{1}-\theta_{2})\,\ddot{x}_{2}
+ m_{2}(l_{2}+x_{2})\sin(\theta_{1}-\theta_{2})\,\ddot{\theta}_{2} \\[4pt]
&\quad
- (m_{1}+m_{2})(l_{1}+x_{1})\,\dot{\theta}_{1}^{2}
+ m_{2}(l_{2}+x_{2})\!\bigl[1+\cos(\theta_{1}-\theta_{2})\bigr]\dot{\theta}_{2}^{2} \\[4pt]
&\quad
+ 2m_{2}\sin(\theta_{1}-\theta_{2})\,\dot{\theta}_{2}\dot{x}_{2}
+ (m_{1}+m_{2})g\cos\theta_{1}
+ k_{1}x_{1}
+ b_{1}\dot{x}_{1}=0
\end{aligned}


$$
\text{Euler–Lagrange Equation for }x_2
$$
\begin{aligned}
&m_{2}\,\ddot{x}_{2}
+ m_{2}\cos(\theta_{1}-\theta_{2})\,\ddot{x}_{1}
- m_{2}(l_{1}+x_{1})\sin(\theta_{1}-\theta_{2})\,\ddot{\theta}_{1} \\[4pt]
&\quad
- m_{2}(l_{1}+x_{1})\cos(\theta_{1}-\theta_{2})\,\dot{\theta}_{1}^{2}
+ m_{2}(l_{2}+x_{2})\,\dot{\theta}_{2}^{2}
- 2m_{2}\sin(\theta_{1}-\theta_{2})\,\dot{\theta}_{1}\dot{x}_{1} \\[4pt]
&\quad
+ m_{2}g\cos\theta_{2}
+ k_{2}x_{2}
+ b_{2}\dot{x}_{2}=0
\end{aligned}


## Symmetry and assumptions

---

#### Time Translation Symmetry

The Lagrangian does not depend explicitly on time:

$$
\frac{\partial \mathcal{L}}{\partial t} = 0
$$

This symmetry implies **conservation of total mechanical energy**.

If I choose no damping in my setup, energy should be conserved over time.

---

#### 1. **Point Masses and Massless Springs**
- The masses $m_1$ and $m_2$ are modeled as **point particles**.
- The springs are **massless**, contributing nothing to the kinetic energy.

#### 2. **Motion Restricted to a Plane**
- The entire system moves within a **2D vertical plane**.

#### 3. **Damping in Radial Direction only**
- I symbolically derive the system including radial damping but simulate the undamped case by setting $ b_1 = b_2 = 0 $ in the final numerical integration and plots.

#### 4. **No Friction or Air Resistance**
- No damping or drag forces are included in the model.


````python
import numpy as np
import sympy as smp
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from numpy.linalg import norm
from matplotlib.animation import FuncAnimation, PillowWriter
from matplotlib.lines      import Line2D
from matplotlib.transforms import Affine2D
````

````python
t, m1, m2, g, k1, k2, l1, l2, b1, b2 = smp.symbols('t m1 m2 g k1 k2 l1 l2 b1 b2')

the1, the2, r1, r2 = smp.symbols(r'\theta_1, \theta_2, r_1, r_2', cls=smp.Function)

# theta1
the1 = the1(t)
the1_d = smp.diff(the1, t)
the1_dd = smp.diff(the1_d, t)

the2 = the2(t)
the2_d = smp.diff(the2, t)
the2_dd = smp.diff(smp.diff(the2, t), t)

r1 = r1(t)
r1_d = smp.diff(r1, t)
r1_dd = smp.diff(smp.diff(r1, t), t)

r2 = r2(t)
r2_d = smp.diff(r2, t)
r2_dd = smp.diff(smp.diff(r2, t), t)

x1, y1, x2, y2 = smp.symbols('x_1, y_1, x_2, y_2', cls=smp.Function)
x1= x1(the1, r1)
y1= y1(the1, r1)
x2= x2(the1, r1, the2, r2)
y2= y2(the1, r1, the2, r2)

x1 = (l1+r1)*smp.sin(the1)
y1 = -(l1+r1)*smp.cos(the1)
x2 = (l1+r1)*smp.sin(the1) + (l2+r2)*smp.sin(the2)
y2 = -(l1+r1)*smp.cos(the1) - (l2+r2)*smp.cos(the2)

# Kinetic energy: two masses
T = (
    smp.Rational(1,2)*m1*(smp.diff(x1, t)**2 + smp.diff(y1, t)**2)
  + smp.Rational(1,2)*m2*(smp.diff(x2, t)**2 + smp.diff(y2, t)**2)
)

# Potential energy: gravity + two springs
V = (
    m1*g*y1
  + m2*g*y2
  + smp.Rational(1,2)*k1*r1**2
  + smp.Rational(1,2)*k2*r2**2
)

# Lagrangian
L = T - V

R = smp.Rational(1,2)*b1*r1_d**2 + smp.Rational(1,2)*b2*r2_d**2
````

````python
LE1 = smp.diff(L, the1) - smp.diff(smp.diff(L, the1_d), t)
LE1 = LE1.simplify()

LE2 = smp.diff(L, the2) - smp.diff(smp.diff(L, the2_d), t)
LE2 = LE2.simplify()

LE3 = (
    smp.diff(L, r1)
  - smp.diff(smp.diff(L, r1_d), t)
  - smp.diff(R, r1_d)
)
LE3 = smp.simplify(LE3)

LE4 = (
    smp.diff(L, r2)
  - smp.diff(smp.diff(L, r2_d), t)
  - smp.diff(R, r2_d)
)
LE4 = smp.simplify(LE4)
````

````python
sols = smp.solve([LE1, LE2, LE3, LE4], (the1_dd, the2_dd, r1_dd, r2_dd),
                simplify=False, rational=False)
````

````python
# Define parameter order and state variables for lambdify
args = (
    m1, m2, g,
    k1, k2,
    l1, l2,
    b1, b2,
    the1, the2, r1, r2,
    the1_d, the2_d, r1_d, r2_d
)

# Lambdify second-derivative expressions
dw1dt_f = smp.lambdify(args, sols[the1_dd], 'numpy')
dw2dt_f = smp.lambdify(args, sols[the2_dd], 'numpy')
dv1dt_f = smp.lambdify(args, sols[r1_dd],    'numpy')
dv2dt_f = smp.lambdify(args, sols[r2_dd],    'numpy')

# Lambdify trivial time-derivatives (identity)
dthe1dt_f = smp.lambdify((the1_d,), the1_d, 'numpy')
dthe2dt_f = smp.lambdify((the2_d,), the2_d, 'numpy')
dr1dt_f   = smp.lambdify((r1_d,),   r1_d,   'numpy')
dr2dt_f   = smp.lambdify((r2_d,),   r2_d,   'numpy')
````

````python
def double_spring_pendulum_derivatives(t, state):
    """
    Compute time‐derivatives for the Rayleigh‐damped double spring pendulum.

    Parameters
    ----------
    t : float
        Current time (unused, but required by ODE solvers).
    state : array_like, shape (8,)
        [theta1, omega1, theta2, omega2, x1, v1, x2, v2]

    Returns
    -------
    dstate_dt : list of float
        [dθ1/dt, dω1/dt, dθ2/dt, dω2/dt, dx1/dt, dv1/dt, dx2/dt, dv2/dt]
    """
    θ1, ω1, θ2, ω2, r1, v1, r2, v2 = state

    # trivial first‐derivative mappings
    dθ1_dt = ω1
    dθ2_dt = ω2
    dr1_dt = v1
    dr2_dt = v2

    # pack common parameters
    params = (m1, m2, g,
              k1, k2,
              l1, l2,
              b1, b2,
              θ1, θ2, r1, r2,
              ω1, ω2, v1, v2)

    # accelerations from the damped EL equations
    dω1_dt = dw1dt_f(*params)
    dω2_dt = dw2dt_f(*params)
    dv1_dt = dv1dt_f(*params)
    dv2_dt = dv2dt_f(*params)

    return [dθ1_dt, dω1_dt, dθ2_dt, dω2_dt, dr1_dt, dv1_dt, dr2_dt, dv2_dt]
````

````python
t = np.linspace(0, 60, 3001)

# 2) Numeric parameter values (do NOT reuse the symbol names)
m1_val, m2_val = 1.0, 1.0
g_val = 9.8
k1_val, k2_val = 10.0, 10.0
l1_val, l2_val = 0.5, 0.5
b1_val, b2_val = 0.0, 0.0

# 3) Initial state vector [θ1, ω1, θ2, ω2, x1, v1, x2, v2]
y0 = [np.pi/4, 0.0,
      np.pi/4, 0.0,
      0.5, 0.0,
      0.5, 0.0]

# 4) Wrapper so odeint sees f(y,t)
def dydt(y, t):
    θ1, ω1, θ2, ω2, x1, v1, x2, v2 = y

    # first‐derivative pieces
    dθ1_dt = ω1
    dθ2_dt = ω2
    dx1_dt = v1
    dx2_dt = v2

    # pack all numeric + state into one tuple, in the same order you used when lambdifying
    params = (
        m1_val, m2_val, g_val,
        k1_val, k2_val,
        l1_val, l2_val,
        b1_val, b2_val,
        θ1, θ2, x1, x2,
        ω1, ω2, v1, v2
    )

    # compute the four accelerations
    dω1_dt = dw1dt_f(*params)
    dω2_dt = dw2dt_f(*params)
    dv1_dt = dv1dt_f(*params)
    dv2_dt = dv2dt_f(*params)

    return [dθ1_dt, dω1_dt,
            dθ2_dt, dω2_dt,
            dx1_dt, dv1_dt,
            dx2_dt, dv2_dt]

def dydt_ivp(t, y):
    return dydt(y, t)

sol = solve_ivp(dydt_ivp, (t[0], t[-1]), y0, t_eval=t,
                    method='RK45', rtol=1e-8, atol=1e-8)

ans = sol.y.T
````

````python
def get_positions(theta1, theta2, x1_ext, x2_ext):
    r1 = l1_val + x1_ext
    r2 = l2_val + x2_ext
    x1 = r1*np.sin(theta1)
    y1 = -r1*np.cos(theta1)
    x2 = x1 + r2*np.sin(theta2)
    y2 = y1 - r2*np.cos(theta2)
    return x1, y1, x2, y2

x1, y1, x2, y2 = get_positions(
    ans[:,0],  # θ1 array
    ans[:,2],  # θ2 array
    ans[:,4],  # x1_ext array
    ans[:,6]   # x2_ext array
)
````

## Vector Field for Four Independent Coords.

````python
def rhs(th1, th1d, th2=0, th2d=0,
        r1=0, r1d=0, r2=0, r2d=0):
    """return (θ̈1, θ̈2, r̈1, r̈2) at the chosen state"""
    p = (m1_val, m2_val, g_val,
         k1_val, k2_val,
         l1_val, l2_val,
         b1_val, b2_val,
         th1, th2, r1, r2,
         th1d, th2d, r1d, r2d)
    return dw1dt_f(*p), dw2dt_f(*p), dv1dt_f(*p), dv2dt_f(*p)

N = 25
th1_lim = (-np.pi, np.pi);  w1_lim = (-4, 4)
th2_lim = (-np.pi, np.pi);  w2_lim = (-4, 4)
r1_lim  = (-1.0, 1.0);      r1d_lim = (-5, 5)
r2_lim  = (-1.0, 1.0);      r2d_lim = (-5, 5)

fig, ax = plt.subplots(2, 2, figsize=(9, 7))

# θ1 – θ̇1
TH1, W1 = np.meshgrid(np.linspace(*th1_lim, N),
                      np.linspace(*w1_lim , N))
dW1 = np.empty_like(TH1)
for i in range(N):
    for j in range(N):
        dW1[i, j], *_ = rhs(TH1[i, j], W1[i, j])
ax[0, 0].quiver(TH1, W1, W1, dW1, color='crimson', angles='xy', lw=1.2)
ax[0, 0].set_xlabel(r'$\theta_1$'); ax[0, 0].set_ylabel(r'$\dot{\theta}_1$'); ax[0, 0].grid()

# θ2 – θ̇2
TH2, W2 = np.meshgrid(np.linspace(*th2_lim, N),
                      np.linspace(*w2_lim , N))
dW2 = np.empty_like(TH2)
for i in range(N):
    for j in range(N):
        _, dW2[i, j], *_ = rhs(0, 0, TH2[i, j], W2[i, j])
ax[0, 1].quiver(TH2, W2, W2, dW2, color='royalblue', angles='xy', lw=1.2)
ax[0, 1].set_xlabel(r'$\theta_2$'); ax[0, 1].set_ylabel(r'$\dot{\theta}_2$'); ax[0, 1].grid()

# r1 – ṙ1
R1, R1d = np.meshgrid(np.linspace(*r1_lim,  N),
                      np.linspace(*r1d_lim, N))
dR1d = np.empty_like(R1)
for i in range(N):
    for j in range(N):
        *_, dR1d[i, j], _ = rhs(0, 0, 0, 0, R1[i, j], R1d[i, j])
ax[1, 0].quiver(R1, R1d, R1d, dR1d, color='darkorange', angles='xy', lw=1.2)
ax[1, 0].set_xlabel(r'$r_1$'); ax[1, 0].set_ylabel(r'$\dot{r}_1$'); ax[1, 0].grid()

# r2 – ṙ2
R2, R2d = np.meshgrid(np.linspace(*r2_lim,  N),
                      np.linspace(*r2d_lim, N))
dR2d = np.empty_like(R2)
for i in range(N):
    for j in range(N):
        *_, _, dR2d[i, j] = rhs(0, 0, 0, 0, 0, 0, R2[i, j], R2d[i, j])
ax[1, 1].quiver(R2, R2d, R2d, dR2d, color='seagreen', angles='xy', lw=1.2)
ax[1, 1].set_xlabel(r'$r_2$'); ax[1, 1].set_ylabel(r'$\dot{r}_2$'); ax[1, 1].grid()

fig.tight_layout()
plt.show()
````

## Energy vs. Time for Leapfrog and RK45

````python
import numpy as np
import matplotlib.pyplot as plt

# Time array
t = np.linspace(0, 60, 3001)
dt = t[1] - t[0]

# Parameters
m1_val, m2_val = 1.0, 1.0
g_val = 9.8
k1_val, k2_val = 10.0, 10.0
l1_val, l2_val = 0.5, 0.5
b1_val, b2_val = 0.0, 0.0

# Initial state [θ1, ω1, θ2, ω2, x1, v1, x2, v2]
y0 = [np.pi/4, 0.0, np.pi/4, 0.0, 0.5, 0.0, 0.5, 0.0]

# Placeholder acceleration functions (should be replaced by actual lambdified functions)
def dw1dt_f(*args): return -g_val / l1_val * np.sin(args[9])
def dw2dt_f(*args): return -g_val / l2_val * np.sin(args[10])
def dv1dt_f(*args): return -k1_val * args[11] / m1_val
def dv2dt_f(*args): return -k2_val * args[13] / m2_val

# RK45 solution (already provided by user)
def dydt(y, t):
    θ1, ω1, θ2, ω2, x1, v1, x2, v2 = y
    dθ1_dt = ω1
    dθ2_dt = ω2
    dx1_dt = v1
    dx2_dt = v2
    params = (
        m1_val, m2_val, g_val,
        k1_val, k2_val,
        l1_val, l2_val,
        b1_val, b2_val,
        θ1, θ2, x1, x2,
        ω1, ω2, v1, v2
    )
    dω1_dt = dw1dt_f(*params)
    dω2_dt = dw2dt_f(*params)
    dv1_dt = dv1dt_f(*params)
    dv2_dt = dv2dt_f(*params)
    return [dθ1_dt, dω1_dt, dθ2_dt, dω2_dt, dx1_dt, dv1_dt, dx2_dt, dv2_dt]

rk_sol = np.zeros((len(t), len(y0)))
rk_sol[0] = y0
for i in range(1, len(t)):
    rk_sol[i] = rk_sol[i-1] + dt * np.array(dydt(rk_sol[i-1], t[i-1]))

# Leapfrog solver
def leapfrog(t_array, y0, dt):
    n_steps = len(t_array)
    y = np.zeros((n_steps, len(y0)))
    y[0] = y0
    θ1, ω1, θ2, ω2, x1, v1, x2, v2 = y0

    for i in range(n_steps - 1):
        params = (
            m1_val, m2_val, g_val,
            k1_val, k2_val,
            l1_val, l2_val,
            b1_val, b2_val,
            θ1, θ2, x1, x2,
            ω1, ω2, v1, v2
        )
        a_θ1 = dw1dt_f(*params)
        a_θ2 = dw2dt_f(*params)
        a_x1 = dv1dt_f(*params)
        a_x2 = dv2dt_f(*params)

        ω1 += 0.5 * dt * a_θ1
        ω2 += 0.5 * dt * a_θ2
        v1 += 0.5 * dt * a_x1
        v2 += 0.5 * dt * a_x2

        θ1 += dt * ω1
        θ2 += dt * ω2
        x1 += dt * v1
        x2 += dt * v2

        params_new = (
            m1_val, m2_val, g_val,
            k1_val, k2_val,
            l1_val, l2_val,
            b1_val, b2_val,
            θ1, θ2, x1, x2,
            ω1, ω2, v1, v2
        )
        a_θ1 = dw1dt_f(*params_new)
        a_θ2 = dw2dt_f(*params_new)
        a_x1 = dv1dt_f(*params_new)
        a_x2 = dv2dt_f(*params_new)

        ω1 += 0.5 * dt * a_θ1
        ω2 += 0.5 * dt * a_θ2
        v1 += 0.5 * dt * a_x1
        v2 += 0.5 * dt * a_x2

        y[i+1] = [θ1, ω1, θ2, ω2, x1, v1, x2, v2]

    return y

leapfrog_sol = leapfrog(t, y0, dt)

# Total energy function
def total_energy(sol):
    θ1, ω1, θ2, ω2, x1, v1, x2, v2 = sol.T
    r1 = l1_val + x1
    r2 = l2_val + x2
    x1_dot = v1
    y1_dot = -r1 * ω1
    x2_dot = v1 * np.cos(θ1 - θ2) + v2
    y2_dot = -r1 * ω1 * np.cos(θ1) - r2 * ω2 * np.cos(θ2)
    T1 = 0.5 * m1_val * (x1_dot**2 + y1_dot**2)
    T2 = 0.5 * m2_val * ((r1 * ω1)**2 + (r2 * ω2)**2)
    T = T1 + T2
    y1 = -r1 * np.cos(θ1)
    y2 = -r1 * np.cos(θ1) - r2 * np.cos(θ2)
    V = m1_val * g_val * y1 + m2_val * g_val * y2 + 0.5 * k1_val * x1**2 + 0.5 * k2_val * x2**2
    return T + V

# Compute and plot energy
E_rk = total_energy(rk_sol)
E_lf = total_energy(leapfrog_sol)

plt.figure(figsize=(10, 5))
plt.plot(t, E_rk, label='RK45 Energy', lw=1, color='red')
plt.plot(t, E_lf, '--', label='Leapfrog Energy', lw=2, color='blue')
plt.xlabel('Time [s]')
plt.ylabel('Total Energy [J]')
plt.title('Energy Conservation: RK45 vs. Leapfrog')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

## Lyapunov exponent

````python
def dydt_vec(t, z):
    """z = [state , δ]  (16-vector)"""
    y      = z[:8]
    delta  = z[8:]

    # flow at the base point
    f      = dydt(y, t)

    # Jacobian numerically (8×8)   ──> finite differences, ε = 1e-8
    J = np.zeros((8, 8))
    eps = 1e-8
    for k in range(8):
        y_pert       = y.copy()
        y_pert[k]   += eps
        f_pert       = dydt(y_pert, t)
        J[:, k]      = (np.array(f_pert) - np.array(f)) / eps

    # variational equation  δ̇ = J·δ
    delta_dot = J @ delta
    return np.hstack((f, delta_dot))

def lyapunov(y0, T=200, dt=0.01, renorm=10.0, seed=0):
    np.random.seed(seed)
    t0, t_end = 0.0, T
    times     = np.arange(t0, t_end, renorm)
    delta0    = np.random.randn(8);  delta0 /= norm(delta0)

    ln_growth = 0.0
    y = y0.copy()

    for t_start in times:
        z0 = np.hstack((y, delta0))
        sol = solve_ivp(dydt_vec, (t_start, t_start+renorm), z0,
                        max_step=dt, rtol=1e-8, atol=1e-8)
        y      = sol.y[:8, -1]
        delta  = sol.y[8:, -1]

        # growth factor
        g      = norm(delta)
        ln_growth += np.log(g)
        delta /= g            # re-normalise

    return ln_growth / (t_end - t0)

λ1 = lyapunov(y0, T=300, dt=0.02, renorm=5.0)
print(f"Lyapunov exponent ≈ {λ1:.4f}")
````

## 4. Exploration of Symmetries and Conserved quantities

### a. Conservation Laws and Symmetry

The double spring pendulum, when simulated **without damping**, is expected to conserve total mechanical energy due to **time translation symmetry** of the Lagrangian.

To visualize this:
- I plotted the **total energy $E(t)$** for both **Leapfrog** and **RK45** integration schemes.
- The **Leapfrog** method preserves energy nearly **constant** throughout time, confirming the expected conservation.
- In contrast, **RK45** exhibits a **gradual drift** in energy, indicating that it does not exactly preserve symplectic structure.

Therefore, the energy plot highlights that the system's symmetry under time translation leads to conservation of total energy, and numerical integrators like Leapfrog respect that symmetry better.

---

### b. Evidence of Chaos

This undamped double spring pendulum shows strong signs of **chaotic behavior**:
- I computed the **largest Lyapunov exponent**, and the value was approximately **$\lambda_1 \approx 0.6234$**, which is **positive**—a key signature of chaos.
- Even in a limited simulation time window (60 seconds due to computational constraints), the **state space trajectories** start to **densely fill regions**, indicating a **sensitive dependence on initial conditions**.

Thus, the system not only exhibits theoretical potential for chaos, but is numerically observed to be chaotic under realistic simulation parameters.



## 7. Utilization of advanced numerical approaches

To investigate the dynamics of the double spring pendulum, I implemented and compared **two numerical integration techniques**:

- **Runge-Kutta 4(5) (RK45):**
  This adaptive method was used for generating accurate state-space trajectories and smooth animations of the system. It handles stiffness moderately well and provides good accuracy in short to medium time intervals. However, over long times, it fails to preserve energy as you can see in $E(t)$ plot.

- **Leapfrog (Velocity-Verlet):**
  To complement RK45, I implemented a symplectic Leapfrog integrator, which is known to preserve the Hamiltonian structure of conservative systems. This method was used specifically to verify energy conservation in the undamped regime. Unlike RK45, Leapfrog maintains a nearly constant total energy over time, providing physical fidelity for long-term simulations.



#### Runge-Kutta 4(5) Method (RK45)

RK45 is an adaptive step-size integrator based on a combination of fourth- and fifth-order Runge-Kutta methods. It is widely used for general-purpose ODE solving due to its balance between accuracy and efficiency.

The RK45 update rule is built on evaluating intermediate slopes:

$$
y_{n+1} = y_n + \sum_{i=1}^6 b_i k_i
$$

where $ k_i $ are derivative evaluations at intermediate points. The difference between 4th and 5th order estimates provides a built-in error control mechanism. For more detailed information, you can go to the [Wikipedia article on RK45](https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta%E2%80%93Fehlberg_method).

---

#### Leapfrog (Velocity-Verlet) Method

Leapfrog is a second-order symplectic integrator ideal for conservative systems like Hamiltonian dynamics. It excels in long-term energy conservation.

The update scheme is:

$$
\begin{aligned}
v_{n+\frac{1}{2}} &= v_n + \frac{\Delta t}{2} a(x_n) \\
x_{n+1} &= x_n + \Delta t \cdot v_{n+\frac{1}{2}} \\
v_{n+1} &= v_{n+\frac{1}{2}} + \frac{\Delta t}{2} a(x_{n+1})
\end{aligned}
$$

This staggered approach ensures time-reversibility and preserves the symplectic structure, making it especially reliable in Hamiltonian systems. For more detailed information, you can go to the [Wikipedia article on Leapfrog](https://en.wikipedia.org/wiki/Leapfrog_integration).

## 10. Hamiltonian vs. Lagrangian Approaches

#### a. Comparative Implementation

For this project, the double spring pendulum was formulated and simulated using **both Lagrangian and Hamiltonian mechanics**.

- **Lagrangian Method**:
  We defined the Lagrangian as $ L = T - V $, where $ T $ is the total kinetic energy and $ V $ is the total potential energy (gravitational and spring-based). The generalized coordinates used were $ \theta_1, \theta_2, r_1, r_2 $, and we derived equations of motion via the Euler-Lagrange equation:
  $$
  \frac{d}{dt}\left( \frac{\partial L}{\partial \dot{q}_i} \right) - \frac{\partial L}{\partial q_i} = 0
  $$
  These second-order ODEs were then solved numerically (e.g., with RK45 or Leapfrog).

- **Hamiltonian Method**:
  We transformed the Lagrangian system into the Hamiltonian formalism:
  $$
  H = \sum_i p_i \dot{q}_i - L
  $$
  where the conjugate momenta are $ p_i = \frac{\partial L}{\partial \dot{q}_i} $. The Hamiltonian $ H $ was expressed in terms of $ q_i $ and $ p_i $, and the equations of motion were derived using Hamilton's equations:
  $$
  \dot{q}_i = \frac{\partial H}{\partial p_i}, \quad \dot{p}_i = -\frac{\partial H}{\partial q_i}
  $$
  The state space was then integrated numerically using `solve_ivp`.

#### b. Benefits and Physical Intuition

- The **Lagrangian approach** offers a more direct connection to energy-based modeling, especially for complex systems with generalized coordinates. It is intuitive for deriving equations of motion without explicitly introducing forces.

- The **Hamiltonian method** has several theoretical and computational advantages:

1. **Phase Space Analysis**
   Hamiltonian mechanics naturally formulates dynamics in terms of positions and momenta.
   This gives you complete control over phase space, which is critical for analyzing chaotic systems, stability, and conservation laws.

2. **Energy Conservation + Symplectic Integrators**
   If you care about long-term numerical accuracy in a conservative system, Hamiltonian-compatible integrators like Leapfrog or Verlet preserve energy better than generic ODE solvers.
   In your project, this was a clear win for the Hamiltonian approach over RK45.

3. **Generalization to Quantum Mechanics**
   Hamiltonian formalism translates cleanly into quantum mechanics, especially in operator and phase-space approaches.
   So if you’re doing classical systems as stepping stones to quantum ones, Hamiltonians are the gateway.


While the Hamiltonian framework is theoretically elegant, I genuinely don’t see any practical benefit in using it for classical mechanics problems like this one expect making phase space plots. The Lagrangian method was already sufficient and more direct for modeling and simulation. The added complexity of converting to canonical variables felt unnecessary in this context.

    0.5,      # r2
    0.0,      # p1

    0.0,      # pr1
    0.0       # pr2
]

def hamiltonian_rhs(t, state):
    θ1, θ2, r1, r2, p1, p2, pr1, pr2 = state
    Δ = θ1 - θ2
    c = np.cos(Δ)

    # radial velocities
    r1d = pr1 / (m1_val + m2_val)
    r2d = pr2 / m2_val

    # invert inertia matrix for angular velocities
    A = (m1_val + m2_val) * r1**2
    B = m2_val * r1 * r2 * c
    C = m2_val * r2**2
    det = A*C - B*B
    θ1d = (C*p1 - B*p2) / det
    θ2d = (A*p2 - B*p1) / det

    # potential derivatives
    dV_dθ1 = (m1_val + m2_val)*g_val * r1 * np.sin(θ1)
    dV_dθ2 = m2_val * g_val * r2 * np.sin(θ2)
    dV_dr1 = -(m1_val + m2_val)*g_val * np.cos(θ1) + k1_val * (r1 - l1_val)
    dV_dr2 = -m2_val * g_val * np.cos(θ2) + k2_val * (r2 - l2_val)

    # momenta derivatives
    dp1  = -dV_dθ1
    dp2  = -dV_dθ2
    dpr1 = -dV_dr1
    dpr2 = -dV_dr2

    return [θ1d, θ2d, r1d, r2d, dp1, dp2, dpr1, dpr2]
